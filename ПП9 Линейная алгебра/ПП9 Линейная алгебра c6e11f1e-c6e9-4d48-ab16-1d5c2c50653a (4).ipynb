```python
# Импортирую необходимые библиотеки
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split

from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score

import matplotlib.pyplot as plt
import seaborn as sb
sb.set_style('darkgrid') # Задаю стиль граффического отбражения данных
#sb.set_context('notebook') # Задаю стиль граффического отбражения данных
```

```python
# создаю класс для работы над проектом.MyLinearRegression. 
# Этот класс обучает линейную регресию,
# делает предсказания,
# шифрует данные,
# показывает параметры(веса и сдвиг) линейной регресии,
# показывает какой обратной матрицей он шифровал данные
class MyLinearRegression:

    def fit(self, train_features, train_target): 
        # fit "обучает" линейную регресию. 
        # train_features - признаки линейно регресии любой размерности
        # train_target - целевой признак
        X = np.concatenate((np.ones((train_features.shape[0], 1)), train_features), axis=1)
        y = train_target
        xtx = np.dot ( X.T , X )
        xtx_obr = np.linalg.inv(xtx)
        xtx_obr_xt = np.dot ( xtx_obr ,  X.T )
        w = np.dot ( xtx_obr_xt , y )
        self.w = w[1:]
        self.w0 = w[0]
        self.ww = w

    def predict(self, test_features):
        # predict "предсказывает" 
        # test_features - признаки по которым делается предсказание
        return test_features.dot(self.w) + self.w0
    
    def crypt(self, features):
        self.obr = np.random.normal( size = (features.shape[1],features.shape[1]))
        return np.dot (features , self.obr )
    
    def param(self):
        return self.w , self.w0
    
    def matr(self):
        return self.obr
```

```python
features = df.drop('Страховые выплаты', axis=1)
target = df['Страховые выплаты']
```

```python
model = MyLinearRegression() # загружаем модель
model.fit(features, target) # обучаем модель-вычисляем аналитически веса и вектор сдвига 
predictions = model.predict(features) # получаем предсказания модели по той же самой выборке признаков
print ( 'r2_score=' , r2_score ( target , predictions ) ) # высичляем метрику r2
w_model , w0_model = model.param() # w_model и w0_model веса и вектор сдвига модели без шифрования
```

```python
model_c = MyLinearRegression() # загружаем модель, _c означает что будет производиться шифроваение данных
features_c = model_c.crypt(features) # преобразуем признаки специально написаным методом
model_c.fit(features_c, target) # обучаем модель-вычисляем аналитически веса и вектор сдвига 
predictions = model_c.predict(features_c)  # получаем предсказания модели по той же самой выборке признаков
print('r2_score=' ,r2_score(target, predictions))  # высичляем метрику r2
w_model_c , w0_model_c = model_c.param()  # w_model_c и w0_model_c веса и вектор сдвига c шифрованием
obr = model_c.matr() # Обратимая матрица, или ключ для расшифровки признаков 
```
